```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### INSURANCE EXAMPLE

```python
#Trial
df=pd.read_csv(r"C:\Users\SAHIL\Downloads\insurance_data.csv")
df.head()
```

```python
plt.scatter(df.age,df.bought_insurance,marker='*',color='blue')
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df[['age']],df['bought_insurance'],train_size=0.8,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
model.score(xtest,ytest)
```

```python
plt.scatter(df.age,df.bought_insurance,marker='*',color='blue')
x1=np.linspace(df['age'].min(),df['age'].max(),100).reshape(-1,1)
prob=model.predict_proba(x1)[:,1]
plt.plot(x1, prob, color='purple')
plt.xlabel("Age")
plt.ylabel("Insurance")
plt.title("AGE VS INSURANCE")
```
